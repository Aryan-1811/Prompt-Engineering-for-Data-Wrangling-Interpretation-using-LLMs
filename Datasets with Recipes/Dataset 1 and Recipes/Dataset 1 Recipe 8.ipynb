```python
#DATA PREPROCESSING
placement_dataset = read.csv('../input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv')
placement_dataset$gender = factor(placement_dataset$gender,
                                 levels = c("M","F"),
                                 labels = c("0","1"))
placement_dataset$ssc_b = factor(placement_dataset$ssc_b,
                                 levels = c("Others","Central"),
                                 labels = c("0","1"))
placement_dataset$hsc_s = factor(placement_dataset$hsc_s,
                                 levels = c("Commerce","Science","Arts"),
                                 labels = c("0","1","2"))
placement_dataset$hsc_b = factor(placement_dataset$hsc_b,
                                 levels = c("Others","Central"),
                                 labels = c("0","1"))
placement_dataset$degree_t = factor(placement_dataset$degree_t,
                                 levels = c("Comm&Mgmt","Sci&Tech","Others"),
                                 labels = c("0","1","2"))
placement_dataset$workex = factor(placement_dataset$workex,
                                 levels = c("No","Yes"),
                                 labels = c("0","1"))
placement_dataset$specialisation = factor(placement_dataset$specialisation,
                                 levels = c("Mkt&HR","Mkt&Fin"),
                                 labels = c("0","1"))
placement_dataset$status = factor(placement_dataset$status,
                                 levels = c("Not.Placed","Placed"),
                                 labels = c("0","1"))
placement_dataset$status = factor(placement_dataset$status,
                                 levels = c("Not.Placed","Placed"),
                                 labels = c("0","1"))
placement_dataset$salary = ifelse(is.na(placement_dataset$salary),
                                 ave(placement_dataset$salary, FUN = function(x)mean (x,na.rm=T)),
                                     placement_dataset$salary)
```

```python
#SPLITTING DATA
library(caTools)
set.seed(1001)
split = sample.split(placement_dataset$mba_p , SplitRatio = 0.80)
train_set = subset(placement_dataset, split=T)
test_set = subset(placement_dataset, split=T)
```

```python
#BUILDING MODEL
regressor = lm(formula = mba_p ~ ssc_p + hsc_p , data = train_set)
summary(regressor)
```

```python
#PREDICTIONS
y_predict = predict(regressor , newdata = test_set)
y_predict
```
