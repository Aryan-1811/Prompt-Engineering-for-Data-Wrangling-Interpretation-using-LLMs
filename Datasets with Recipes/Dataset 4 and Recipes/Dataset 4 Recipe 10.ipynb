```python
data = pd.read_csv('/kaggle/input/real-estate-price-prediction/Real estate.csv')
data
```

```python
data.info()
```

```python
## There are no null entries here
## All columns are numerical
```

```python
data = data.drop(['No'], axis = 1)
data
```

```python
## Now, renaming the columns to easily understand names
data['X1 transaction date'].unique()
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=101)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
## Train this model on Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
## from coef and intercept we can make a equation also
## The equation will look like this
```

## $real estate price = -12815.542+5.06*(X1 transaction date)-0.2836*(X2 house age)-0.004814*(X3 distance to the nearest MRT station)+1.0026*(X4 number of convenience stores)+230.62*(X5 latitude)--25.4512(X6 longitude)$

```python
## Evaluation of the model
## we will calculate first MAE, that will take sum of absolute difference of
## predicted_output and original_output
sum=0
for i in range(len(y_pred)):
    sum=sum + abs(y_pred[i]-y_test[i])

mae = sum/len(y_pred)
print(mae)
```
